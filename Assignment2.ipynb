```python
import numpy as np
import pandas as pd
from IPython.display import display,HTML


class Titanic_data:

    def header(self,header,font=("Aptos",20,"bold")):
        display(HTML(f"<h2 style='text-align:center;color:green'>{header} </h2>"))
    
    def lheader(self,header,font=("aptos",8)):
        display(HTML(f"<h2 style='text-align:left;color:green'>{header} </h2>"))
        

    def static(self,data):
        self.header("Load the Titanic dataset")
        display(data)
        self.header("display the first 10 rows")
        display(data.head(10))
        self.header("Display the last 10 rows")
        display(data.tail(10))
        self.header("Find the number of rows and columns")
        display(data.shape)
        self.header("Display the columns names")
        display(data.columns.to_list())
        self.header("Display the data types af all columns")
        display(data.info())  
        self.header("Generate summary statistics of numerical columns")
        display(data.describe(include='number'))
        self.header("Generate summary statistics of categorical columms")
        display(data.describe(include='object'))
        self.header("Display random 15 records")
        display(data.sample(15))
        self.header("Identify numerical and categorical columns") 
        self.lheader("Numerical columns")
        display(data.select_dtypes(include='number').columns.to_list())
        self.lheader("Categorical columns")
        display(data.select_dtypes(include='object').columns.to_list())


    def Data_selection(self,data):
        self.header("Display only the Name columns")
        display(data.loc[:,'Name'])
        self.header("display Name,Age and Fare")
        display(data.loc[:,['Name','Age','Fare']])
        self.header("display passengers age is greater then 40 ")
        display(data[data['Age']>40])
        self.header("display female passengers")
        display(data[data['Sex']=='female'])
        self.header("display passengers who survived")
        display(data['Survived'])
        self.header("display passengers paid a fare greater then 100")
        display(data[data['Fare']>100])
        self.header("Display passengers traveling in First class")
        display(data[data['Pclass']==1])
        self.header("Age is between 20 to 40 ")
        display(data[(data['Age']>=20) & (data['Age']<=40)])
        self.header("Embarked from southampton(s)")
        display(data[data['Embarked'] == 'S'])
        self.header("passengers are younger then 18")
        display(data[data['Age']>18])

    def Data_Cleaning(self,data):
        self.header("Find the missing values in each columns")
        missing = data.isnull().sum()
        display(missing)
        self.header("Identify the column with the highest number of missing values")
        display(missing[missing == missing.max()])
        self.header("Display all rows where Age is missing")
        display(data[data['Age'].isnull()])
        self.header("display All rows where cabin is missing")
        display(data[data['Cabin'].isna()])
        self.header("Remove rows where Age is missing")
        display(data.dropna(subset=['Age']))
        self.header("Fill missing Age values using the median")
        data['Age'].fillna(data['Age'].median(),inplace=True)
        display(data['Age'].isnull().sum())
        self.header("Fill missing Embarked values using the mode")
        data['Embarked'].fillna(data['Embarked'].mode(),inplace=True)
        display(data['Embarked'].isnull().sum())
        self.header("Replace missing cabin values with unknown")
        display(data['Cabin'].fillna("Unknown",inplace=True))
        self.header("Replace value ")
        display(data['Sex'].replace({
            'male' : 'Male',
            'female' : 'Female'
        }))
        self.header("Columns Rename")
        display(data.rename(columns={
            'Pclass' : 'Passenger_Class',
            'Fare' : 'Ticket_Price'
        },inplace=True))
        self.header("Check whether duplicate records exits")
        display(data.duplicated().sum())
        self.header("Remove duplicate records")
        display(data.drop_duplicates(inplace=True))
        self.header("Verify that duplicates have been removed")
        display(data.duplicated().sum())
        self.header("Save the cleaned dataset ")
        data.to_csv("Titanic_Cleaned.csv",index=False)
        display(pd.read_csv("Titanic_Cleaned.csv"))

    def Data_Manipulation(self,data):
        self.header("Create columns Family_Size")
        data['Family_Size'] = data['SibSp'] + data['Parch']
        display(data['Family_Size'])
        self.header("Create column Total_Memers")
        data['Total_member'] = data['SibSp']+data['Parch']+1
        # self.header("Create Fare_Per_Person")
        # data['Fare_Per_Person'] = data['Fare']/data['Total_member']
        # display(data['Fare_Per_Person'])
        self.header("create columns passengers_type")
        def passenger_type(Age):
            if Age<18:
                return "Child"
            elif Age<60:
                return "Adult"
            else:
                return "Senior"
        data['Passenger_Type'] = data['Age'].apply(passenger_type)
        display(data['Passenger_Type'].sample(5))    
        self.header("Increase all ticket prices by 10%")
        data['Fare'] = data['Fare'] * 1.10
        display(data['Fare'])
        self.header("Give an additional 5% increase only to first class passengers")
        data['Fare'] = data['Fare'][data['Pclass'] == 1]*1.05
        display(data['Fare'])
        self.header("Delete the Ticket column")
        data.drop(columns='Ticket',inplace=True)
        display(data.head())
        self.header("Sort Passengers by ticket price (highest first)")
        display(data.sort_values(by="Fare",ascending=False))
        self.header("Sort passengers by age(youngest first)")
        display(data.sort_values(by='Age'))
        self.header("sort by passenger class and ticket price")
        display(data.sort_values(by=['Pclass','Fare']))

        
    def str_opeation(self,data):
        self.header("Convert all passenger names to uppercase")
        display(data['Name'].str.upper())
        self.header("convert all passenger names to lowercase")
        display(data['Name'].str.lower())
        self.header("Find all passengers names contain 'Mr'")
        display(data[data['Name'].str.contains('Mr',na=False)])
        self.header("passengers name contain 'Mrs'")
        display(data[data['Name'].str.contains('Mrs',na=False)])
        self.header("passengers name contain 'Miss'")
        display(data[data['Name'].str.contains('Miss',na=False)])
        self.header("Replace 'Mr.  with 'Mister' ")
        data['Name'] = data['Name'].replace('Mr.','Mister',regex=False)
        display(data)
        

    def Aggregation_func(self,data):
        self.header("Total Fare collected")
        display(data['Fare'].sum())
        self.header("Average Fare")
        display(data['Fare'].mean())
        self.header("Median Fare")
        display(data['Fare'].median())
        self.header("Minimum Fare")    
        display(data['Fare'].min())
        self.header("Maximum Fare")
        display(data['Fare'].max())
        self.header("Standard deviation of fare")
        display(data['Fare'].std())
        self.header("count total passengers")
        display(data['Fare'].count())
        self.header("Average passenger age")
        display(data['Age'].mean())
        self.header("Highest ticket price")
        display(data['Fare'].max())
        self.header("Lowest ticket price")
        display(data['Fare'].min())


    def Groupby_operation(self,data):
        self.header("count passengers in each passenger class")
        display(data.groupby(by='Pclass')['Pclass'].count())
        self.header("count passengers by gender")
        display(data.groupby(by='Sex')['Sex'].count())
        self.header("calculate average fare by passenger class")
        display(data.groupby(by='Pclass')['Fare'].mean())
        self.header("claculate average age by passenger class")
        display(data.groupby(by='Pclass')['Age'].mean())
        self.header("calculate maximum fare by passenger class")
        display(data.groupby(by='Pclass')['Fare'].max())
        self.header("calculate minimum fare by passenger class")
        display(data.groupby(by='Pclass')['Fare'].min())
        self.header("calculate toal fare collected from each passenger class")
        display(data.groupby(by='Pclass')['Fare'].sum())
        self.header("Find survival count by gender")
        display(data.groupby(by='Survived')['Sex'].count())
        self.header("find average fare by gender")
        display(data.groupby(by='Sex')['Fare'].mean())
        self.header("Generate a report showing")
        display(data.groupby(by='Pclass')['Fare'].agg(['count','mean','median','min','max','std']))
        

    def Merge_operation(self,data):
        insurance = pd.DataFrame({
            "PassengerId" : [1,2,3,5,8,900],
            "Policy" : ["Gold","Silver","Gold","Permimum","Silver","Gold"]
        })

        meal = pd.DataFrame({
            "PassengerId" : [1,2,4,5,7,900],
            "Meanl" : ["Veg","Non-Veg","Veg","Veg","Non-Veg","Veg"]
        })
        self.header("Inner Merge between Titanic and insurance")
        display(pd.merge(
            data,
            insurance,
            on = 'PassengerId',
            how = 'inner'
        ))
        self.header("Left Merge")
        display(pd.merge(
            data,
            insurance,
            on='PassengerId',
            how='left'
        ))
        self.header("Right Merge")
        display(pd.merge(
            data,
            insurance,
            on = 'PassengerId',
            how = 'right'
        ))
        self.header("Outer Merge")
        display(pd.merge(
            data,
            insurance,
            on='PassengerId',
            how='outer'
        ))
        self.header("identify passengers who do not have insurance")
        display(data['PassengerId'][~data['PassengerId'].isin(insurance['PassengerId'])])
        self.header("identify insurance records that do not have matching passengers")
        display(data[~data['PassengerId'].isin(insurance['PassengerId'])])


    def Concatenation(self,data):
        jan = pd.DataFrame({
            "PassengerId" : [1001,1002,1003],
            "Passenger_Name" : ["Aman","Riya","Karan"]
        })
        feb = pd.DataFrame({
            "PassengerId" : [1004,1005],
            "Passenger_Name" : ["Neha","Rahul"]
        })
        city = pd.DataFrame({
            "City" : ['Delhi','Mumbai','Pune','Jaipur','Lucknow']
        })
        self.header("vertical concatenation of jan & feb bookings")
        display(pd.concat([jan,feb],axis=0))
        self.header("perform vertical concatenation using ignore_index=True")
        display(pd.concat([jan,feb],axis=0,ignore_index=False))
        self.header("horizontal concatenation of booking and city")
        display(pd.concat([jan,feb,city],axis=1))

    def date_time(self,data):
        self.header("create a new column named Booking_date")
        data['Booking_Date'] = pd.to_datetime(
                                        np.random.choice(
                                        pd.date_range('2024-01-01', '2024-12-31'),
                                        size=len(data)
                                        ))
        display(data.head())
        self.header("conver into datetime formate")
        data['Booking_Date'] = pd.to_datetime(data['Booking_Date'])
        display(data.head())
        self.header("Extract")
        display("Year")
        display()
                
        
        
        


data = pd.read_csv("titanic.csv")
obj = Titanic_data()
```
