# Concate

```python
df1 = pd.DataFrame({
    "Emp_id" : [101,102],
    "Name" : ['Vinay','Aman']
})
df2 = pd.DataFrame({
    "Emp_id" : [103,104],
    "Name" : ['Riya','Priya']
})
```

```python
# by default axis=0 
vertical = pd.concat([df1,df2])
vertical
```

```python
vertical = pd.concat([df1,df2],axis=1)
vertical
```

```python
employee = pd.DataFrame({
    "Emp_id" : [101,102,103],
    "Name" : ['Vinay','Aman','Riya']
})

salary = pd.DataFrame({
    "Salary" : [50000,45000,60000],
    "Bonus" : [5000,3000,7000]
})

pd.concat([employee,salary],axis=1)
```

# Marge

```python
employee = pd.DataFrame({
    "Emp_id" : [101,102,103,104],
    "Name" : ['Vinay','Aman','Riya','Priya']
})

salary = pd.DataFrame({
    "Emp_id" : [101,102,105],
    "Salary" : [50000,45000,60000]
})
```

```python
inner = pd.merge(
    employee,
    salary,
    on = "Emp_id",
    how = "right"
)
print(inner)
```
