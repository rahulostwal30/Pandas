```python
from IPython.display import display,HTML
import numpy as np
import pandas as pd

class Student:
    
    def header(self, header,font=None):
        display(HTML(f"<h2 style='text-align:center;color:green;'>{header}</h2>"))

    def static(self, data):
        self.header("Load the Dataset",font=("times of roman",20,"bold"))
        display(data)
        self.header("Display the first 10 Rows",font=("times of roman",20,"bold"))
        display(data.head(10))
        self.header("Display the last 10 Rows",font=("times of roman",20,"bold"))
        display(data.tail(10))
        self.header("Find the number of rows and columns",font=("times of roman",20,"bold"))
        display(data.shape)
        self.header("Display all columns",font=("times of roman",20,"bold"))
        display(data.columns.to_list)
        self.header("Check the data types of every columns",font=("times of roman",20,"bold"))
        display(data.dtypes)
        self.header("Display the dataset summery",font=("times of roman",20,"bold"))
        display(data.info)
        self.header("Display Descriptive statistics",font=("times of roman",20,"bold"))
        display(data.describe)
        self.header("check the memory usage",font=("times of roman",20,"bold"))
        display(data.memory_usage)

    def exploring(self,data):
        self.header("Dataset contains missing values",font=("times of roman",20,"bold"))
        display(data.isnull().any()) 
        self.header("count missing values column-wise",font=("times of roman",20,"bold"))
        display(data.isnull().sum())
        self.header("Check duplicate rows",font=("times of roman",20,"bold"))
        display(data.duplicated().sum())
        self.header("Remove duplicate Rows",font=("times of roman",20,"bold"))
        display(data.drop_duplicates())
        self.header("Count unique values in every columns",font=("times of roman",20,"bold"))
        display(data.nunique())
        self.header("Display unique gender",font=("times of roman",20,"bold"))
        display(data['gender'].unique())
        self.header("Display unique parental education levels",font=("times of roman",20,"bold"))
        display(data['parental level of education'].unique())
        self.header("Display unique lunch",font=("times of roman",20,"bold"))
        display(data['lunch'].unique())
        self.header("Display unique test preparation categories",font=("times of roman",20,"bold"))
        display(data['test preparation course'].unique())
        self.header("Display the frequency of each gender",font=("times of roman",20,"bold"))
        display(data['gender'].value_counts())


    def selecting_data(self,data):
        self.header("Display the math score column",font=("times of roman",20,"bold"))
        display(data['math score'])
        self.header("Display the reading & writting score",font=("times of roman",20,"bold"))
        display(data[['reading score','writing score']])
        self.header("Display the first 20 Rows",font=("times of roman",20,"bold"))
        display(data[:20])
        self.header("Display the Rows 30-50",font=("times of roman",20,"bold"))
        display(data[30:51])
        self.header("Display the only categorical column",font=("times of roman",20,"bold"))
        display(data.select_dtypes(include='object').columns.to_list)
        self.header("Display the only numerical column",font=("times of roman",20,"bold"))
        display(data.select_dtypes(include='number').columns.to_list)
        self.header("Display the first five students math score ",font=("times of roman",20,"bold"))
        display(data['math score'][:5])
        self.header("Display the last five student's reading score",font=("times of roman",20,"bold"))
        display(data['reading score'][-5:])
        self.header("select column using .loc",font=("times of roman",20,"bold"))
        display(data.loc[0:5,'gender':'lunch'])
        self.header("select rows using .iloc",font=("times of roman",20,"bold"))
        display(data.iloc[0:5,0:3])


    def filtering(self,data):
        self.header("Students with math score above 90",font=("times of roman",20,"bold"))
        display(data[data['math score']>90])
        self.header("Students with reading score above 80",font=("times of roman",20,"bold"))
        display(data[data['reading score']>80])
        self.header("Students with writin score above 90",font=("times of roman",20,"bold"))
        display(data[data['writing score']>40])
        self.header("Female Students",font=("times of roman",20,"bold"))
        display(data[data['gender']=='female'])
        self.header("Male students",font=("times of roman",20,"bold"))
        display(data[data['gender']=='male'])
        self.header("Students who completed the test preparation course",font=("times of roman",20,"bold"))
        display(data[data['test preparation course']=='compeleted'])
        self.header("Students who received standard lunch",font=("times of roman",20,"bold"))
        display(data[data['lunch']=='standard'])
        self.header("Students whose parents have a bachelor's degree",font=("times of roman",20,"bold"))
        display(data[data['parental level of education']=="bachelor's degree"])
        self.header("Students with math score above 90 and reading score above 90",font=("times of roman",20,"bold"))
        display(data[(data['math score']>90) & (data['reading score']>90)])
        self.header("Student whose reading score is greater then writing score")
        display(data[data['reading score']>data['writing score']])
        

    def sorting(self,data):
        self.header("sort students by math score",font=("times of roman",20,"bold"))
        display(data.sort_values(by='math score'))
        self.header("sort by reading score in desc order",font=("times of roman",20,"Bold"))
        display(data.sort_values(by='reading score',ascending=False))
        self.header("sort by writing score",font=("times of roman",20,"Bold"))
        display(data.sort_values(by='writing score'))
        self.header("sort by gender and math score",font=("times of roman",20,"Bold"))
        display(data.sort_values(by=['gender','math score']))
        self.header("Display the top 10 students on math score",font=("times of roman",20,"Bold"))
        display(data.sort_values(by='math score',ascending=False).head(10))
    
    def create(self,data):
        self.header("Create a Total score column",font=("times of roman",20,"bold"))
        data['Total Score'] = data[['math score','reading score','writing score']].sum(axis=1)
        display(data)
        self.header("Create a Average score column",font=("times of roman",20,"bold"))
        data['Average'] = data[['math score','reading score','writing score']].mean(axis=1)
        display(data)
        self.header("Create a Result column",font=("times of roman",20,"bold"))
        data['Result'] = np.where(data['Average']>=40,"Pass","Fail")
        display(data)
        self.header("Create a Greade column",font=("times of roman",20,"bold"))
        data['Grade'] = data['Average'].map(lambda x : "A" if x>=90
                                                      else "B" if x>=80
                                                      else "C" if x>=70
                                                      else "D" if x>60
                                                      else "F"
                                            )
        display(data)
        self.header("create a performance columns",font=("times of roman",20,"bold"))
        data['Performance'] = data['Average'].map(lambda x : "Excellent" if x>=90
                                                            else "Good" if x>=75
                                                            else "Average" if x>=60
                                                            else "Poor"
                                                  )
        display(data)


    def statis(self,data):
        self.header("Mean math score",font=("times of roman",20,"bold"))
        display(data['math score'].mean())
        self.header("Median reading score",font=("times of roman",20,"bold"))
        display(data['reading score'].median())
        self.header("Standard deviation of writing score",font=("times of roman",20,"bold"))
        display(data['writing score'].std())
        self.header("Highest math score",font=("times of roman",20,"bold"))
        display(data['math score'].max())
        self.header("Lowest reading score",font=("times of roman",20,"bold"))
        display(data['reading score'].min())
        self.header("Range of writing score",font=("times of roman",20,"bold"))
        display(data['writing score'].max() - data['writing score'].min())
        self.header("Correlation between math and reading",font=("times of roman",20,"bold"))
        display(data['math score'].corr(data['reading score']))
        self.header("correlation among all numerical columns",font=("times of roman",20,"bold"))
        display(data.corr(numeric_only=True))
        self.header("Variance",font=("times of roman",20,"bold"))
        display(data['math score'].var())
        self.header("calculate the percentiles",font=("times of roman",20,"bold"))
        display("25th => ",data['math score'].quantile(0.25))
        display("50th => ",data['math score'].quantile(0.50))
        display("75th => ",data['math score'].quantile(0.75))


    def groupby(self,data):
        self.header("Average math score by gender",font=("times of roman",20,"bold"))
        display(data.groupby(by='gender')['math score'].mean())
        self.header("Average reading score by gender",font=("times of roman",20,"bold"))
        display(data.groupby(by='gender')['reading score'].mean())
        self.header("Average writing score by gender",font=("times of roman",20,"bold"))
        display(data.groupby(by='gender')['writing score'].mean())
        self.header("Average scores based on lunch type",font=("times of roman",20,"bold"))
        display(data.groupby(by='lunch')[['math score','reading score','writing score']].mean())
        self.header("Average scores based on parental education",font=("times of roman",20,"bold"))
        display(data.groupby(by='parental level of education')
                [['math score','reading score','writing score']].mean())
        self.header("Average scores based on preparation course",font=("times of roman",20,"bold"))
        display(data.groupby(by='test preparation course')
                [['math score','reading score','writing score']].mean())
        self.header("Count students in each race/ethnicity.",font=("times of roman",20,"bold"))
        display(data['race/ethnicity'].value_counts())
        self.header("Highest math score in every race",font=("times of roman",20,"bold"))
        display(data.groupby(by='race/ethnicity')['math score'].max())
        self.header("Lowest reading score by gender",font=("times of roman",20,"bold"))
        display(data.groupby(by='gender')['reading score'].min())
        self.header("count students by parental education",font=("times of roman",20,"bold"))
        display(data['parental level of education'].value_counts())
        

    def value_count(self,data):
        self.header("count each gender",font=("times of roman",20,"bold"))
        display(data['gender'].value_counts())
        self.header("count lunch categories",font=("times of roman",20,"bold"))
        display(data['lunch'].value_counts())
        self.header("count test preparation categories",font=("times of roman",20,"bold"))
        display(data['test preparation course'].value_counts())
        self.header("count parental education levels",font=("times of roman",20,"bold"))
        display(data['parental level of education'].value_counts())
        self.header("count race/ethnicity groups",font=("times of roman",20,"bold"))
        print(data['race/ethnicity'].value_counts())


    def real_world(self,data):
        pass

    def bounus_challenge(self,data):
        self.header("score above 90 in all three subjects",font=("times of roman",20,"bold"))
        display(data[(data['math score']>90) & 
                     (data['reading score']>90) & 
                     (data['writing score']>90)])
        self.header("scored below 40 in all three subject",font=("times of roman",20,"bold"))
        display(data[(data['math score']<40) &
                     (data['reading score']<40) &
                     (data['writing score']<40)])
        self.header("Percentage of students who completed the test preparation course"
                    ,font=("times of roman",20,"bold"))
        display(((data['test preparation course']
                    [data['test preparation course']=='completed'].count())
                    /data.shape[0])*100)
        self.header("Highest-performing race/ethinicity group based on average total score",
                    font=("times of roman",20,"bold"))
        new = data['race/ethnicity'].value_counts()
        new = new[new > new.mean()]
        display(new[new == new.max()])
        self.header("which subject [Math,Reading,Writing ] has the highest average score",
                    font=("times of roman",20,"bold"))
        h_avg = data[['math score','reading score','writing score']].mean()
        display(h_avg[h_avg == h_avg.max()])
        self.header("Whose Reading score is at least 20 marks higher then their math score",
                    font=("times of roman",20,"bold"))
        display(data[data['reading score'] > data['math score']+20])
        self.header("Create Merit List sorted by total score",
                    font=("times of roman",20,"bold"))
        Merit_List = data.sort_values(by='Total Score',ascending=False)
        display(Merit_List)
        self.header("Identify the top 5 students in each subject",
                    font=("times of roman",20,"bold"))
        display(data.sort_values(by=['math score','reading score','writing score']
                                 ,ascending=False).iloc[:5])
        self.header("Export the final processed dataset",font=("times of roman",20,"bold"))
        display(data.to_csv("students_performance_cleaned.csv",index=False))
        
        
        



    

data = pd.read_csv("StudentsPerformance.csv")   # Dataset
obj = Student()
```
